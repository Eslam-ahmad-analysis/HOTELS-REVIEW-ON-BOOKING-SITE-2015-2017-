```python
import pandas as pd
import numpy as np
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
from tqdm import tqdm
import time

# ----------- إعداد الجيوكودر -------------
geolocator = Nominatim(user_agent="my_geo_app", timeout=10)

geocode = RateLimiter(
    geolocator.geocode,
    min_delay_seconds=1.5,
    max_retries=3,
    error_wait_seconds=3,
    swallow_exceptions=False
)

# ----------- Caching -------------
cache = {}
errors = []

def get_coords_cached(address):
    """
    ترجّع (lat, lng) للعنوان باستخدام Caching + RateLimiter
    """
    if pd.isnull(address):
        return (np.nan, np.nan)
    
    # لو العنوان موجود أصلاً في الكاش
    if address in cache:
        return cache[address]
    
    try:
        location = geocode(address)
        if location:
            coords = (location.latitude, location.longitude)
        else:
            coords = (np.nan, np.nan)
    except Exception as e:
        # تسجّل العنوان اللي فشل
        errors.append(f"{address} --> {str(e)}")
        coords = (np.nan, np.nan)
    
    # نخزن في الكاش
    cache[address] = coords
    return coords

# ----------- استهداف الصفوف اللي ناقصاها lat/lng -------------
mask = df['lat'].isna() | df['lng'].isna()
missing_addresses = df.loc[mask, 'hotel_address']

# Progress bar
tqdm.pandas(desc="Geocoding addresses")

# ----------- تطبيق الدالة على العناوين الناقصة فقط -------------
results = missing_addresses.progress_apply(
    lambda addr: pd.Series(get_coords_cached(addr))
)

# ----------- كتابة النتيجة في الداتا -------------
df.loc[mask, ['lat', 'lng']] = results.values

# ----------- حفظ الأخطاء لو موجودة -------------
if errors:
    with open("geocoding_errors.txt", "w", encoding="utf-8") as f:
        for err in errors:
            f.write(err + "\n")
```
